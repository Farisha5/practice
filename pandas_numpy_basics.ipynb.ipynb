```python
import pandas as pd
import numpy as np
```

```python
data = {
    "Sample_ID": [1, 2, 3, 4, 5],
    "Gene_Expression": [2.3, 3.1, np.nan, 4.0, 3.5],
    "Condition": ["Control", "Treated", "Control", "Treated", "Control"]
}

df = pd.DataFrame(data)
df
```

```python
treated_samples = df[df["Condition"] == "Treated"]
treated_samples
```

```python
df_filled = df.fillna(df["Gene_Expression"].mean())
df_filled
```

```python
expression_values = np.array(df_filled["Gene_Expression"])
expression_values
```

```python
print("Mean:", np.mean(expression_values))
print("Max:", np.max(expression_values))
```
