```python
import matplotlib.pyplot as plt

genes = ['GeneA', 'GeneB', 'GeneC', 'GeneD']
expression = [5.2, 3.8, 7.1, 2.5]

plt.plot(genes, expression, marker='o', linestyle='-', color='blue')
plt.title('Gene Expression Levels')
plt.xlabel('Genes')
plt.ylabel('Expression')
plt.show()
```

```python
plt.bar(genes, expression, color='green')
plt.title('Gene Expression Bar Chart')
plt.xlabel('Genes')
plt.ylabel('Expression')
plt.show()
```

```python
plt.hist(expression, bins=4, color='orange', edgecolor='black')
plt.title('Distribution of Gene Expression')
plt.xlabel('Expression Levels')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.bar(genes, expression)
plt.savefig('gene_expression_bar.png')   # Saves in current folder
plt.close()  # Closes the plot
```
