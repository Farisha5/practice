```python
import pandas as pd
import matplotlib.pyplot as plt

# Example dataset
data = {
    'Gene': ['GeneA', 'GeneB', 'GeneC', 'GeneD', 'GeneE'],
    'Expression': [5.2, 3.8, 7.1, 2.5, 6.0]
}

df = pd.DataFrame(data)
print(df)

# Line Plot
plt.plot(df['Gene'], df['Expression'], marker='o')
plt.title('Line Plot of Gene Expression')
plt.savefig('line_plot.png')
plt.close()

# Bar Chart
plt.bar(df['Gene'], df['Expression'], color='purple')
plt.title('Bar Chart of Gene Expression')
plt.savefig('bar_chart.png')
plt.close()

# Histogram
plt.hist(df['Expression'], bins=5, color='skyblue', edgecolor='black')
plt.title('Histogram of Gene Expression')
plt.savefig('histogram.png')
plt.close()
```
